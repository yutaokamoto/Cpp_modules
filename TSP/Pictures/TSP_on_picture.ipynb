````python
import cv2
import os
import numpy as np
import random

img = cv2.imread("lyricalschool.png",cv2.IMREAD_COLOR)
#print(img[150, 30])
#print(len(img))
#print(len(img[1]))

C={}
demand={}
c_counter=0
#写真の座標を見て、黒だったら顧客リストに座標追加
for i in range(len(img)):
    for j in range(len(img[0])):
        iro = sum(list(img[i,j])) 
        r=random.randint(0,iro) 
        if iro==0:
            r=random.randint(0,1) #1/20の確率で顧客ノード追加
        if r==0:
            C["c"+str(c_counter)] = (i,j)
            demand["c"+str(c_counter)]=1
            c_counter += 1
C["depot"] = C["c"+str(len(C)-1)]
print("顧客数",len(C))
````

````python
import time
start = time.time()
opt_val=0

not_done=[i for i in C]
now="c0"
not_done.remove(now)
tsp_list=[now]
tsp_dict={}

for i in range(len(not_done)):
    dis=big_m
    for i in not_done:
       
        if distance(C[now],C[i])<dis:
            dis = distance(C[now],C[i])
            next_node = i
    opt_val += distance(C[now],C[next_node])
    
    not_done.remove(next_node)
    tsp_dict[now]=next_node
    now=next_node
    tsp_list.append(next_node)
tsp_dict[now]="c0"
tsp_list.append("c0")
elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
print("コスト",opt_val)
````

# 2_opt

````python
#２本の枝を交換して、交換した後整える関数
def two_opt_swap (a,c,tsp_dict,C):
    #わかりやすくするためにa,b,c,d
    b=tsp_dict[a]
    d=tsp_dict[c]
    
    two_opt_num=0 #2optが行われたか判定
    #交換したらコストが安くなるか
    if (distance(C[a],C[b])+distance(C[c],C[d]))>(distance(C[a],C[c])+distance(C[b],C[d])):
        
        tsp_dict[a] = c#つなぎ変え
        
        before_node=b#ここからcまでの経路を反転させる
        now_node=tsp_dict[b] 
        
        tsp_dict[b] = d#つなぎ変え
        
        #枝の向きを反転させる
        while True:
            next_node=tsp_dict[now_node] 
            tsp_dict[now_node]=before_node
            before_node=now_node
            now_node = next_node
            two_opt_num=1
            if before_node==c:
                break
                
    return tsp_dict ,two_opt_num
````

## ２_opt法　Neighborバージョン

````python
#結局交換が行われるのは近くの点なので、近傍の10個 くらい見ときゃ大丈夫で、１００個も見れば十分らしい
````

````python
#neighbor使ったバージョン

## neighbor に近い順のリストを入れる
import time
start = time.time()
neighbor={}
for i in C:
    neighbor_distance_list=[(distance(C[i],C[j]),j) for j in C if i!=j]
    neighbor_distance_list.sort()
    neighbor[i]=[i[1] for i in neighbor_distance_list]
    
elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
````

````python
#2_opt開始
start = time.time()

zz=0
while True:
    print(zz)
    zz+=1
    
    end_num=0#「改善できなくなったら終わる」ために導入したもの
    for i in C:
        for j in neighbor[i][0:101]:#近傍100個（50個でも結果は変わらなかった）
            tsp_dict,tow_opt_num=two_opt_swap(i,j,tsp_dict,C)
            end_num+=tow_opt_num#改善ができたらend_numに1が足されていく
    if end_num==0:
        break

elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")


#コスト計算
two_opt_val = 0
for i in tsp_dict:
    edge_list.append((i,tsp_dict[i]))
    two_opt_val+=distance(C[i],C[tsp_dict[i]])

print("コスト=",two_opt_val)
````

````python
G2 = nx.Graph()
two_opt_val=0#2_optの巡回路の距離
for i in C:
    G2.add_node(i)
edge_list=[]
for i in tsp_dict:
    edge_list.append((i,tsp_dict[i]))
    two_opt_val+=distance(C[i],C[tsp_dict[i]])
G2.add_edges_from(edge_list)
fig=plt.figure(figsize=(13,13 ))
nx.draw_networkx(G2,pos=C,nodelist=C,node_color='black',node_size=0.1,with_labels=False,edge_color="k",width=1)
plt.show()
````
