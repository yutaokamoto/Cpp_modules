```python
class Node:
    def __init__(self,name):
        self.name=name
        self.right=name*10#右の子の名前
        self.left=name*10+1#左の子の名前　
        self.depth=len(str(name))-1#木の深さ　最大でアイテム数
        self.value= self.value_func()#現在割り当てられてるアイテムの価値の合計（ナップサックに入りきっていなくても）
        self.room=self.room_func() #空き容量
        self.relax= self.simplex_func()+self.value#緩和解
        
        
    def room_func(self):#現在のナップサックの残り容量
            room=K#ナップサックの容量　　クラス外で定義　←これは良くない？
            Assigned=[int(v) for i,v in enumerate(list(str(self.name))) if i>0]#今入ってるアイテムに対応したリスト ex[1,0,1]ならアイテム１、３が入ってる
            #self.name=101ならAssignedaは[1,0,1]
            for i,v in enumerate(Assigned):
                if v==1:
                    room -= W[i]#←クラス外で定義したW←これは良くない？
            return room
        
    def value_func(self):#現在割り当てられてるアイテムの価値の合計（ナップサックに入りきっていなくても）
        value=0
        Assigned=[int(v) for i,v in enumerate(list(str(self.name))) if i>0]#今入ってるアイテムに対応したリスト ex[1,0,1]ならアイテム１、３が入ってる
        #self.name=101ならAssignedaは[1,0,1]
        for i,v in enumerate(Assigned):
            if v==1:
                value += V[i]#←クラス外で定義したW←これは良くない？
        return value
    
    
    def simplex_func(self):
        x = 0
        Capacity = self.room
        Unassigned=list(range(self.depth,N))
        Value_per_Weight_order = sorted(Unassigned, key=lambda i: V[i]/W[i], reverse=True) #アイテムの並べ替え これはリスト
        for i in Value_per_Weight_order:
            if W[i]<=Capacity:
                x += V[i]
                Capacity -= W[i]
            else:
                x += V[i]*(Capacity/W[i])
                #return x
        return x
a=Node(1)
```

```python
now = Node(1)
stack = []
stack.append(now)
opt_temp = 0
opt_solution_temp=0
while stack!=[]:
     #スタックの一番上のアイテムを取り出す
    now = stack.pop()
    print(now.name)
    #そのノードにおける解が実行不可能
    if now.room>=0:
        #緩和解が今の解よりいい
        if now.relax > opt_temp:
            ##子がある
            if now.depth<N:
                ###スタックに子を入れる
                stack.append(Node(now.right))
                stack.append(Node(now.left))

            ##子がない→葉になる 緩和解と実際の解が同じになる
            else:
                opt_temp = now.value
                opt_solution_temp=now.name
                print("opt_temp=",opt_temp)     
                continue
        else:
            continue
    
    else:
        continue
    
    
        
        
    
    print("opt_temp=",opt_temp)            

        
        
    print()
print()
print("OPTVAL=",opt_temp)
print(opt_solution_temp)
```
